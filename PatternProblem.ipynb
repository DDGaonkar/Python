##### Pattern 1

```python
for i in range(1, 6): #rows
    for j in range(1, i+1): #columns
        print(j, end=" ")
    print()
```

##### Pattern 2

```python
for i in range(1, 6): #rows
    for j in range(6, i, -1): #columns
        print(j, end=" ")
    print()
```

##### Pattern 3

```python
for i in range(1, 6):
    for j in range(5, i, -1):
        print(" ", end=" ")
    for k in range(i):
        print("*", end=" ")
    print()
```

##### Pattern 4

```python
for i in range(1, 6):
    for j in range(i, 0, -1):
        print(j, end=" ")
    print()
```

##### Pattern 5

```python
for i in range(1, 6):
    for j in range(1, i+1):
        print("*", end=" ")
    print()
for i in range(1, 5):
    for j in range(5, i, -1):
        print("*", end=" ")
    print()
```

##### Pattern 6

```python
for i in range(1, 11):
    for j in range(1, i+1):
        print(i*j, end=" ")
    print()
```
